```python
import matplotlib.pyplot as plt
gs=x['type']
g=x['Alexandria, Egypt']
figureObject, axesObject = plt.subplots()
axesObject.pie(g,startangle=90)
axesObject.legend(gs,loc='center left', bbox_to_anchor=(1.0, 0.5))
axesObject.axis('equal')
plt.show()
plt.tight_layout()
```
